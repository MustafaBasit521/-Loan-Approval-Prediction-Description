# Importing the Libraries

```python
# import libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

# Data Loading


```python
df=pd.read_csv("Loan_approval_prediction.csv")
df.head()
```

# Data Preprocessing

```python
# Data Preprocessing
# Handle categorical variables

# Handle missing values
for col in ["person_income","person_home_ownership","loan_intent","cb_person_default_on_file"]:
    df[col].fillna(df[col].mode()[0], inplace=True)

X = df[["person_income","person_home_ownership","loan_intent","cb_person_default_on_file"]]
X = pd.get_dummies(X, drop_first=True)
y = df["loan_status"]
```

# Splitting the data into Train-Test

```python
# train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,stratify=y, random_state=42)
```

# Feature Scaling

```python
# feature scaling
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

# Model Training

```python
# model training
l_r=LogisticRegression(class_weight='balanced')
d_tree=DecisionTreeClassifier(class_weight='balanced')
```

# Model Fitting


```python
# model fitting
l_r.fit(X_train, y_train)
d_tree.fit(X_train, y_train)
```

# Model Prediction


```python
# model prediction
y_pred_l_r = l_r.predict(X_test)
y_pred_d_tree = d_tree.predict(X_test)
```

# Model - Evaluation Metrics

```python
# Evaluation-Metrics
# Visualizing the Model and comparing the logistic Regression and Decision Tree
print("Logistic Regression Accuracy:", accuracy_score(y_test, y_pred_l_r))
print("Decision Tree Accuracy:", accuracy_score(y_test, y_pred_d_tree))
print("Logistic Regression Classification Report:\n", classification_report(y_test, y_pred_l_r))
print("Decision Tree Classification Report:\n", classification_report(y_test, y_pred_d_tree))
print("Logistic Regression Confusion Matrix:\n", confusion_matrix(y_test, y_pred_l_r))
print("Decision Tree Confusion Matrix:\n", confusion_matrix(y_test, y_pred_d_tree))
```
